## **A Lunar Lander agent that will learn to land correctly on the moon**

Our goal is to train an agent that can land a lunar module on the moon. The agent will receive a reward based on how well it lands, and our objective is to maximize this reward.

We'll be using **[Stable_Baselines3](https://stable-baselines3.readthedocs.io/en/master/)**, a Deep Reinforcement Learning library that provides implementations of various RL algorithms. 

The environment we'll be using is the **[LunarLander-v2](https://gymnasium.farama.org/environments/box2d/lunar_lander/)** environment from OpenAI's Gym. In this environment, the agent controls a lunar lander and must learn to land it safely on the moon's surface. 

Also, to generate a replay video we'll need to have a virtual screen to be able to render the environment. For that:
